```python
import pandas as pd
df = pd.read_csv("cs_research_paper_dataset")
df
```

```python
df = df[['abstract', 'title']]
df
```

```python
df["AbstractLength"] = df["abstract"].str.split(' ').str.len()
df
```

```python
df["id"] = df.index
df
```

```python
dataset1 = df[df["id"] < 30000]
dataset2 = df[(df["id"] >= 30000) & (df["id"] < 60000)]
dataset3 = df[df["id"] >= 60000]
```

```python
dataset1.to_csv("dataset1.csv")
dataset2.to_csv("dataset2.csv")
dataset3.to_csv("dataset3.csv")
```
