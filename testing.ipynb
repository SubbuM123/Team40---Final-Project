```python
import recsys_ii
import inv_index

import numpy as np
import nltk
import pandas as pd
from nltk.corpus import stopwords

from collections import Counter
import math
import time
```

```python
target_idx = 57917
ks = [1, 1.2, 1.4, 1.5, 1.6, 1.8]
vocab_sizes = [500, 1000, 1500]
for k in ks:
    ri = inv_index.InvInd(bm25_k = k, top_words=1500)
    ri.preprocess_data("abstract")
    ri.build_vocab("abstract")
    ri.compute_IDF("abstract")
    ri.preprocess_data("title")
    ri.build_vocab("title")
    ri.compute_IDF("title")
    ri.inverted_index()
    print("Done with ii, starting testing")
    for v in vocab_sizes:
        rs = recsys_ii.RecSys_II(bm25_k=k, top_words=v)
        rs.build_vocab("abstract")
        rs.build_vocab("title")
        qt = "Implicit semantic text retrieval and distributed implementation for rural medical care"
        ss = rs.similarity_ranking(qt, qt)
        top100_indices = np.argsort(ss)[-100:][::-1]

        if target_idx in top100_indices:
            rank = np.where(top100_indices == target_idx)[0][0] + 1  # Rank is 1-based index
            reciprocal_rank = 1 / rank
            with open("hp.txt", "a") as f:
                f.write("k: " + str(k) + "\n")
                f.write("vocab size: " + str(v) + "\n")
                f.write("MRR: " + str(reciprocal_rank) + "\n")
                f.write("-----------------------------------------------------------------------\n")
    print("Done with the iteration")

```
